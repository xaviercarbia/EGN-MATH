*Matrix Version Damped Harmonic Oscillator of a Yo-Yo

```python
model = np.array([[0,1],[-2,-0.1]])
```

```python
def matprint(mat, fmt="g"):
    col_maxes = [max([len(("{:"+fmt+"}").format(x)) for x in col]) for col in mat.T]
    for x in mat:
        for i, y in enumerate(x):
            print(("{:"+str(col_maxes[i])+fmt+"}").format(y), end="  ")
        print("")
```

```python
matprint(model)
```

```python
plt.matshow(model)
```

The Plot above shows the different vibrations intensity characterized by color that the Yo-Yo performs as it does different tricks.

```python
def dx_dt(x,t):
    return model @ x 
```

```python
d = 2
x = np.array([0,1])

t  = 0
tf = 50
dt = 0.001

n = int((tf-t)/dt)

X = np.zeros((n,d))
```

d stands for number of cycles the Yo-Yo is doing and t stands for the total of seconds the Yo-Yo is doing the cycles for.

```python
for i in range(n):

    x = x + dx_dt(x,t)*dt

    X[i] = x
```

```python
plt.plot(X[:,0]);
```

This graph shows the intensity of the up and down motions vibrations of the cycles of the Yo-Yo. 

*Predator-Prey System of Sharks and Minnows

```python
def dx_dt(x,t):
    return np.array([
                      alpha*x[0]  - beta*x[0]*x[1],
                                   delta*x[0]*x[1] - gamma*x[1]  
    ])
```

```python
d = 2
x = np.zeros(d,)

x[0] = 10
x[1] = 10

t  = 0
tf = 50
dt = 0.001

n = int((tf-t)/dt)
X = np.zeros((n,d))
```

d stands for the number of rounds in the sharks and minnows game, x[0] and x[1] stand for the number of sharks and minnows. t stands for the number of seconds in each round.

```python
alpha = 1 
beta = 0.05 
delta = 0.02
gamma = 0.5
```

```python
for i in range(n):

    x = x + dx_dt(x,t)*dt

    X[i] = x
```

```python
plt.plot(X)
```

This graph showed the sharks in blue and minnows in orange. It also showed that the sharks overpowered the minnows because of the powerful rate and advantage that they had to catch the minnows because all they had to do was touch the minnows.

```python
rho = 27.0
sigma = 10.0
beta = 7.0 / 2.0
```

rho stands for number of weeks the weather was tracked, sigma stands for the number of people tracked these weather patterns for these number of weeks, beta stands for number of days the weather was harshest divided by the two most exteme days of the year.

```python
model = np.array([[-sigma,sigma,0,0,0],[rho,-1,0,0,-1],[0,0,-beta,1,0]])
```

```python
matprint(model)
```

```python
plt.matshow(model)
```

This plot shows the different colors that represent the conditions of Florida's weather throughout the year. When the color is blue or closest to blue, it means the the weather was stable. The colors that are most opposite of blue means these were the times of the year when the weather was harshest.

```python
def dx_dt(x, t):
    x, y, z = x
    state = np.array([x,y,z,x*y,x*z])
    return model @ state 
```

```python
d = 3
x = np.array([1,1,1])

t  = 0
tf = 50
dt = 0.001

n = int((tf-t)/dt)
X = np.zeros((n,d))
```

d stands for the harshest amount of months in the Florida year, and t stands for the harshest amount of days of the weather inside those months. 

```python
for i in range(n):

    x = x + dx_dt(x,t)*dt

    X[i] = x
```

```python
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(X[:, 0], X[:, 1], X[:, 2]);
```

This system is an end result of the different change in weather patterns in Florida in a particular year.
